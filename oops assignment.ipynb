```python
#(1) What is Object-Oriented Programming (OOP)+
#    Object-Oriented Programming (OOP) is a programming paradigm that organizes code into 
#    reusable blueprints called classes and objects.

#    It makes the code more: Modular
#    Reusable
#    Easy to maintain
#    Secure

#(2) What is a class in OOP+
#    A class is a blueprint or template for creating objects. It defines the properties 
#    (attributes) and behaviors (methods) that objects of that class will have.
#    Think of a class as a blueprint for a house, and an object as an actual house built from that blueprint.

#(3) What is an object in OOP+
#    An object is an instance of a class. It is a real-world entity that has attributes (data) and behaviors (methods).
#    Think of a class as a blueprint, and an object as the actual thing built from that blueprint.

#(4) What is the difference between abstraction and encapsulation+
#    Difference Between Abstraction and Encapsulation in OOP
#    Both abstraction and encapsulation are fundamental concepts of Object-Oriented 
#    Programming (OOP), but they serve different purposes.

#       1. Abstraction (Hiding Implementation Details)
#    Definition:         Hiding unnecessary details and showing only the essential features of an object.
#    Purpose:            Focuses on what an object does rather than how it does it.
#    Example:            A car's steering wheel hides the complex mechanics of turning the wheels.
#    Implementation:     Achieved using abstract classes and interfaces.

#       2. Encapsulation (Data Hiding & Security)
#    Definition:        Wrapping data (variables) and methods (functions) into a single unit (class) and restricting access to some details.
#    Purpose:           Protects data from direct modification and ensures controlled access.
#    Example:           Bank account details are private and can only be accessed via specific methods.
#    Implementation:    Achieved using private (__), protected (_), and public access modifiers.

# (5) What are dunder methods in Python+
#    Dunder methods (short for Double UNDERscore methods) are special built-in methods in Python that start and end with double underscores (__). 
#    These methods enable classes to have special behaviors like operator overloading, object representation, iteration, and more.
#    They are also known as "magic methods" because Python calls them automatically in certain situations.

#    __init__(self, ...): This is the constructor method, called when an instance (object) of the class is created.

#    __str__(self): Defines the string representation of an object. This method is called by the built-in str() function and the print() function.

#    __repr__(self): Defines the "official" string representation of an object. It is used by the repr() function and typically intended for debugging.

#    __len__(self): This method is used to define the behavior of the len() function applied to an instance of your class.

#    __getitem__(self, key): Allows access to the items of your class using the indexing syntax (obj[key]).

#    __setitem__(self, key, value): Allows setting items in your class using the indexing syntax (obj[key] = value).

#(6) Explain the concept of inheritance in OOPH
#    Inheritance is one of the key concepts of Object-Oriented Programming (OOP). It allows a child class (subclass)
#    to inherit properties and behaviors (methods) from a parent class (superclass).
#    It helps in code reuse, reducing redundancy, and implementing the DRY (Don't Repeat Yourself) principle.

#    Code Reusability → Write code once and reuse it in multiple classes.
#    Extensibility → Add new features without modifying existing code.
#    Hierarchy Representation → Create relationships between different entities.
#    Polymorphism Support → Allows method overriding.

#    Types of Inheritance in Python
#    Single Inheritance
#    Multiple Inheritance
#    Multilevel Inheritance
#    Hierarchical Inheritance
#    Hybrid Inheritance

#(7)  What is polymorphism in OOP+
#    Polymorphism is a foundational concept in Object-Oriented Programming (OOP) that allows objects of different classes 
#    to be treated as objects of a common superclass. It lets you define a single interface for multiple underlying forms
#    (data types). In simpler terms, polymorphism lets one interface interact with different data types and structures.

#    There are two main types of polymorphism in OOP:
#    Compile-time Polymorphism (or Static Polymorphism):
#    Achieved through method overloading and operator overloading.
#    Method overloading allows a class to have multiple methods with the same name, but different parameter lists.
#    Run-time Polymorphism (or Dynamic Polymorphism):
#    Achieved through method overriding.
#    Method overriding allows a subclass to provide a specific implementation of a method that is already defined in its superclass.

#(8) How is encapsulation achieved in Python+
#    Encapsulation is one of the key principles of Object-Oriented Programming (OOP) that restricts direct access to the 
#    data and methods inside a class. It helps in data hiding and ensures that internal class details are not exposed unnecessarily.
#    Encapsulation is achieved by:
#    Using access modifiers (public, protected, private).
#    Using getter and setter methods to access or modify private data.

#   Modifier	  Syntax	  Access Level
#   Public	      variable	  Accessible from anywhere.
#   Protected	 _variable	  Accessible within the class and subclasses.
#   Private	    __variable	  Not accessible directly outside the class.

#(9)  What is a constructor in Python+
#     A constructor is a special method that gets called automatically when an object of a class is created.
#    It is used to initialize the object's attributes.
#    In Python, the constructor method is defined using __init__().

#(10) What are class and static methods in Python+
#     A class method works with the class itself rather than an instance (object). It is defined using the 
#     @classmethod decorator and takes cls as its first parameter.

#     Key Points:
#     Works at the class level (not instance level).
#     Can modify class variables.
#     Defined using @classmethod.
#     Takes cls (class reference) as the first argument.

#     A static method is just like a regular function, but inside a class. It does not take self or cls as a parameter.
#     It is used when we don't need access to instance attributes (self) or class variables (cls).

#     Key Points:
#     Defined using @staticmethod.
#     Does NOT modify class or instance attributes.
#     Used for utility/helper functions.

#    Feature	                    Class Method	                Static Method
#    Decorator	                    @classmethod	                @staticmethod
#    First Parameter	            cls (class reference)	        No cls or self
#    Accesses Class Variables?   	Yes	                             No
#    Accesses Instance Variables?	 No	                             No
#    Used For	                    Modifying class attributes	   Utility/helper functions

#(11) What is method overriding in OOP+

#    Method Overloading allows multiple methods in a class to have the same name but different numbers or types of parameters.
#    In Python, method overloading is NOT directly supported like in Java or C++, but we can achieve similar functionality
#    using default arguments or *args and **kwargs.

#(12)  What is a property decorator in Python+
#     The @property decorator in Python is used to define getter, setter, and deleter methods for class attributes in an elegant and readable way.
#     It allows us to control access to private attributes while keeping the syntax simple. 

#     Encapsulation – Control access to private variables.
#     Readability – Allows accessing methods like attributes (obj.value instead of obj.get_value()).
#     Validation – Add logic to validate data before setting it.

#    Component	                 Purpose
#    @property	                 Defines a getter method for an attribute.
#    @property_name.setter	     Defines a setter method to modify the attribute.
#    @property_name.deleter	     Defines a deleter method to remove the attribute.

#(13)  What is a property decorator in Python+
#      The @property decorator is used in Python to control access to instance variables without changing 
#      the syntax of attribute access. It allows you to define getter, setter, and deleter methods while treating them as regular attributes.
#      Key Benefits:
#      Encapsulation – Controls access to private attributes.
#      Readability – Allows calling methods as attributes (obj.value instead of obj.get_value()).
#      Validation – Adds logic to check values before assigning them.

#(14) Why is polymorphism important in OOP+
#     Polymorphism means "many forms" and allows a single interface (method, function, or operator) to work with different types of objects.
#     Key Idea: The same method name or operator can behave differently depending on the object calling it.

#     1. Importance of Polymorphism in OOP
#     1. Improves Code Reusability
#     You can use the same method name in different classes without rewriting code.
#     Example: A draw() method can be used in multiple shape classes (Circle, Rectangle, Triangle).
#     2. Enhances Flexibility and Maintainability
#     Makes it easier to modify and extend code without affecting existing functionality.
#     Example: You can add a new shape (e.g., Polygon) without modifying existing code.
#     3. Supports Method Overriding
#     Allows a child class to provide a specific implementation of a method already defined in the parent class.
#     Example: A speak() method in a base class Animal, which is overridden in Dog and Cat classes.
#     4. Enables Operator Overloading
#     Allows operators like +, *, == to behave differently based on the object type.
#     Example: "Hello" + "World" results in "HelloWorld", but 5 + 10 results in 15

#(15) What is an abstract class in Python+
#     An abstract class in Python is a blueprint for other classes. It cannot be instantiated and must be inherited
#     by subclasses that provide concrete implementations of its abstract methods.

#      Key Features:
#      Defined using ABC (Abstract Base Class) from the abc module.
#      Contains abstract methods (methods without implementation).
#      Cannot be instantiated directly.
#      Forces subclasses to implement specific methods.

#(16)  What are the advantages of OOP+
#      Object-Oriented Programming (OOP) offers several advantages that make it a popular paradigm 
#      or software development Here are some key benefits:

#      Modularity:
#      OOP allows you to break down complex problems into smaller, more manageable pieces by creating objects
#      (instances of classes). This modularity makes code easier to understand, maintain, and debug.

#      Reusability:
#      With OOP, you can reuse existing classes and objects in new programs. This reduces redundancy and 
#      improves productivity by allowing you to leverage previously developed code.

#      Inheritance:
#      Inheritance allows you to create new classes based on existing ones. It promotes code reusability and enables 
#      you to extend or modify the behavior of existing classes without rewriting them from scratch.

#      Encapsulation:
#      Encapsulation helps in hiding the internal state and implementation details of an object. It provides a clear 
#      interface for interacting with the object while protecting its integrity. This leads to more secure and robust code.

#      Polymorphism:
#      Polymorphism allows objects of different classes to be treated as objects of a common superclass. It enables 
#      you to write more flexible and extensible code by defining methods that work with objects of different types.

#       Abstraction:
#      Abstraction simplifies complex systems by providing a high-level view and hiding unnecessary details. 
#      It allows you to focus on essential features and behavior while ignoring implementation specifics.

#      Improved Collaboration:
#      OOP promotes a clear structure and organization of code, making it easier for teams to collaborate on 
#      large projects. Each team member can work on different classes or modules without interfering with others' work.

#      Scalability:
#      OOP makes it easier to manage and scale large codebases. By organizing code into classes and objects,
#      you can more easily add new features and functionality without disrupting existing code.

#(17) What is the difference between a class variable and an instance variable+ 
#     Class Variables
#     Definition: Class variables are shared among all instances of a class. They are defined within a class, but outside of any methods.
#     Access: You can access class variables using the class name or an instance, but they are common to all instances.
#     Purpose: They are used for attributes and properties that are shared by all instances.

#      Instance Variables
#      Definition: Instance variables are unique to each instance of a class. They are defined within the constructor (__init__ method
#      or other instance methods.
#      Access: You can access instance variables only through a specific instance.
#      Purpose: They are used for attributes and properties that are specific to each instance.

#(18) What is multiple inheritance in Python+
#     Multiple inheritance allows a class to inherit from more than one parent class. This means a child class can have features
#     of multiple base classes.

#     Key Features:
#     A class can inherit methods & attributes from multiple parent classes.
#     Helps in code reusability but can lead to complexity.
#     Python follows the Method Resolution Order (MRO) to determine which method to execute if there's a conflict.

#(19) Explain the purpose of ‘’__str__’ and ‘__repr__’ ‘ methods in PythonH
#     ’__str__’
#     Python provides special methods (also called dunder methods) to define how objects are represented as strings.
#     the two most commonly used are
#     Used when calling str(object) or print(object).
#     Should return a human-readable string that describes the object in a friendly way.
#     ‘__repr__’
#     Used when calling repr(object).
#     Should return a detailed, unambiguous string that can recreate the object if passed to eval().
#     Primarily for debugging and development.

#(20) What is the significance of the ‘super()’ function in Python+
#     super() in Python is used to call methods from a parent class inside a child class. It helps in code reuse, 
#     especially in inheritance and method overriding.

#     Key Significance of super()
#️    Calls Parent Class Methods → Allows access to inherited methods without explicitly naming the parent class.
#️    Supports Multiple Inheritance → Ensures correct Method Resolution Order (MRO).
#️    Enhances Code Maintainability → Reduces redundancy and improves scalability.

#(21)  What is the significance of the __del__ method in Python+
#     __del__() Method in Python (Destructor)
#     The __del__() method is a special destructor method in Python that is automatically called when an object is deleted or
#     goes out of scope. It is used for cleaning up resources, such as closing files or releasing memory.

#      Key Significance of __del__()
#      Releases Resources → Helps in freeing up system resources like database connections or files.
#      Called Automatically → Python calls __del__() when the object is no longer in use.
#      Used for Cleanup → Can be used to perform last-minute tasks before an object is destroyed.

#(22) What is the difference between @staticmethod and @classmethod in Python+
#     In Python, both @staticmethod and @classmethod are used to define methods inside a class that are not regular
#     instance methods. However, they have key differences in how they access class attributes and methods.

#    1. @staticmethod (Independent Method)
#    Does NOT access the class (cls) or instance (self).
#    Works like a regular function inside a class.
#    Used when the method does not need class-specific data.

#    2. @classmethod (Class-Level Method)
#    Takes cls as the first parameter, which represents the class itself.
#    Can modify class-level attributes but not instance attributes.
#    Useful when the method should be shared across all instances.

# Key Differences
#     Feature	                           @staticmethod	      @classmethod
#     #Can modify class variables	              No	              Yes
#     Can modify  instance variables	          No	               No
#     Works like a regular function	              Yes	               No
#     When to use?	                    When no class/instance  	When modifying class-level attributes
#                                             data is needed

#(23) How does polymorphism work in Python with inheritance+

#    Polymorphism in Python, combined with inheritance, allows objects of different classes to be treated as objects of a common superclass.
#    This means that a single interface can be used to interact with different types of objects, making the code more flexible and extensible. 
#    Here's how polymorphism works with inheritance:
# Example
#     Consider a base class Shape with a method area() that is meant to be overridden by subclasses.
class Shape:
    def area(self):
        raise NotImplementedError("Subclass must implement abstract method")

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * (self.radius ** 2)

# Polymorphism in action
shapes = [Rectangle(4, 5), Circle(3)]

for shape in shapes:
    print(shape.area())  # Output: 20 (for Rectangle), 28.26 (for Circle)
```

```python
#(2) Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle
#    from it and implement the area() method in both.
from abc import ABC, abstractmethod

# Abstract class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass  # Abstract method (must be implemented by child classes)

# Derived class Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2  # πr² formula

# Derived class Rectangle
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width  # l × w formula

# Creating objects
circle = Circle(5)
rectangle = Rectangle(4, 6)

# Calling area() method
print(f"Circle Area: {circle.area()}")      # ✅ Output: Circle Area: 78.5
print(f"Rectangle Area: {rectangle.area()}")  # ✅ Output: Rectangle Area: 24
```

```python
#(3) Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car
#   and further derive a class ElectricCar that adds a battery attribute.
# Base class
class Vehicle:
    def __init__(self, type):
        self.type = type  # General attribute for vehicle type

    def show_type(self):
        print(f"Vehicle Type: {self.type}")

# Derived class Car (inherits from Vehicle)
class Car(Vehicle):
    def __init__(self, type, brand):
        super().__init__(type)  # Call the parent class constructor
        self.brand = brand  # New attribute for Car

    def show_car_info(self):
        print(f"Car Brand: {self.brand}")

# Further derived class ElectricCar (inherits from Car)
class ElectricCar(Car):
    def __init__(self, type, brand, battery_capacity):
        super().__init__(type, brand)  # Call the parent class constructor
        self.battery_capacity = battery_capacity  # New attribute for ElectricCar

    def show_electric_car_info(self):
        print(f"Battery Capacity: {self.battery_capacity} kWh")

# Creating an instance of ElectricCar
tesla = ElectricCar("Electric", "Tesla Model 3", 75)

# Calling methods from all levels
tesla.show_type()               # ✅ Output: Vehicle Type: Electric
tesla.show_car_info()           # ✅ Output: Car Brand: Tesla Model 3
tesla.show_electric_car_info()  # ✅ Output: Battery Capacity: 75 kWh
```

```python
#(5) Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
#    balance and methods to deposit, withdraw, and check balance.

class BankAccount:
    def __init__(self, initial_balance):
        self.__balance = initial_balance  # Private attribute (encapsulation)

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ₹{amount}")
        else:
            print("Invalid deposit amount!")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: ₹{amount}")
        else:
            print("Insufficient balance or invalid amount!")

    def check_balance(self):
        print(f"Current Balance: ₹{self.__balance}")

# Creating an account
account = BankAccount(1000)

# Performing operations
account.deposit(500)    # ✅ Output: Deposited: ₹500
account.withdraw(300)   # ✅ Output: Withdrawn: ₹300
account.check_balance() # ✅ Output: Current Balance: ₹1200

# Attempting to access private attribute (will cause an error)
# print(account.__balance)  ❌ AttributeError: 'BankAccount' object has no attribute '__balance'
```

```python
#(7) Create a class MathOperations with a class method add_numbers() to add two numbers and a static
#     method subtract_numbers() to subtract two numbers.

class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        """Class method to add two numbers"""
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        """Static method to subtract two numbers"""
        return a - b

# Using class method
sum_result = MathOperations.add_numbers(10, 5)
print(f"Sum: {sum_result}")  # ✅ Output: Sum: 15

# Using static method
difference = MathOperations.subtract_numbers(10, 5)
print(f"Difference: {difference}")  # ✅ Output: Difference: 5
```

```python
#(10) Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
#     vectors.
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        """Overloads the + operator to add two vectors"""
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        """Returns the string representation of the vector"""
        return f"({self.x}, {self.y})"

# Creating vector objects
v1 = Vector(2, 3)
v2 = Vector(4, 5)

# Adding vectors using operator overloading
v3 = v1 + v2  # Equivalent to v1.__add__(v2)

# Printing results
print(f"Vector 1: {v1}")  # ✅ Output: Vector 1: (2, 3)
print(f"Vector 2: {v2}")  # ✅ Output: Vector 2: (4, 5)
print(f"Sum of Vectors: {v3}")  # ✅ Output: Sum of Vectors: (6, 8)
```

```python
#(12) Implement a class Student with attributes name and grades. Create a method average_grade() to compute
#     the average of the grades.
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades  # List of grades

    def average_grade(self):
        """Computes and returns the average of the grades"""
        if self.grades:
            return sum(self.grades) / len(self.grades)
        return 0  # Return 0 if no grades are available

# Creating a Student object
student1 = Student("Alice", [85, 90, 78, 92])

# Calling the average_grade method
print(f"{student1.name}'s Average Grade: {student1.average_grade():.2f}")  
# ✅ Output: Alice's Average Grade: 86.25
```

```python
#(13)  Create a class Rectangle with methods set_dimensions() to set the dimensions and area() to calculate the
#      area.
class Rectangle:
    def __init__(self):
        self.length = 0
        self.width = 0

    def set_dimensions(self, length, width):
        """Sets the dimensions of the rectangle"""
        self.length = length
        self.width = width

    def area(self):
        """Calculates and returns the area of the rectangle"""
        return self.length * self.width

# Creating a Rectangle object
rect = Rectangle()

# Setting dimensions
rect.set_dimensions(5, 10)

# Calculating and printing area
print(f"Rectangle Area: {rect.area()}")  # ✅ Output: Rectangle Area: 50
```

```python
#(14). Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
#      and hourly rate. Create a derived class Manager that adds a bonus to the salary
class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        """Computes salary based on hours worked and hourly rate"""
        return self.hours_worked * self.hourly_rate

# Derived class Manager
class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        """Computes salary with an additional bonus"""
        return super().calculate_salary() + self.bonus

# Creating Employee object
emp = Employee("Alice", 40, 20)
print(f"{emp.name}'s Salary: ${emp.calculate_salary()}")  
# ✅ Output: Alice's Salary: $800

# Creating Manager object
mgr = Manager("Bob", 40, 30, 500)
print(f"{mgr.name}'s Salary: ${mgr.calculate_salary()}")  
# ✅ Output: Bob's Salary: $1700
```

```python
#(15) Create a class Product with attributes name, price, and quantity. Implement a method total_price() that
#     calculates the total price of the product.
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        """Calculates and returns the total price of the product"""
        return self.price * self.quantity

# Creating Product objects
product1 = Product("Laptop", 50000, 2)
product2 = Product("Phone", 20000, 3)

# Calculating and printing total price
print(f"Total price of {product1.name}: ₹{product1.total_price()}")  
# ✅ Output: Total price of Laptop: ₹100000

print(f"Total price of {product2.name}: ₹{product2.total_price()}")  
# ✅ Output: Total price of Phone: ₹60000
```

```python
#(16) Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that
#     implement the sound() method.
from abc import ABC, abstractmethod

# Abstract class
class Animal(ABC):
    @abstractmethod
    def sound(self):
        """Abstract method to define animal sound"""
        pass

# Derived class Cow
class Cow(Animal):
    def sound(self):
        return "Moo!"

# Derived class Sheep
class Sheep(Animal):
    def sound(self):
        return "Baa!"

# Creating objects
cow = Cow()
sheep = Sheep()

# Calling the sound method
print(f"Cow Sound: {cow.sound()}")   # ✅ Output: Cow Sound: Moo!
print(f"Sheep Sound: {sheep.sound()}")  # ✅ Output: Sheep Sound: Baa!
```

```python
#(17) Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that
#     returns a formatted string with the book's details
class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        """Returns a formatted string with book details"""
        return f"'{self.title}' by {self.author}, published in {self.year_published}"

# Creating Book objects
book1 = Book("The Great Gatsby", "F. Scott Fitzgerald", 1925)
book2 = Book("1984", "George Orwell", 1949)

# Printing book details
print(book1.get_book_info())  
# ✅ Output: 'The Great Gatsby' by F. Scott Fitzgerald, published in 1925

print(book2.get_book_info())  
# ✅ Output: '1984' by George Orwell, published in 1949
```

```python
#(18) Create a class House with attributes address and price. Create a derived class Mansion that adds an
#      attribute number_of_rooms.
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

    def get_details(self):
        """Returns house details"""
        return f"Address: {self.address}, Price: ₹{self.price}"

# Derived class Mansion
class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

    def get_details(self):
        """Returns mansion details including number of rooms"""
        return f"{super().get_details()}, Rooms: {self.number_of_rooms}"

# Creating House object
house = House("123 Main St, Delhi", 5000000)
print(house.get_details())  
# ✅ Output: Address: 123 Main St, Delhi, Price: ₹5000000

# Creating Mansion object
mansion = Mansion("456 Luxury Rd, Mumbai", 20000000, 10)
print(mansion.get_details())  
# ✅ Output: Address: 456 Luxury Rd, Mumbai, Price: ₹20000000, Rooms: 10
```
